```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from tensorflow.keras.utils import to_categorical

df = pd.read_csv( 'ph-data.csv')

# last column is the target variable
x = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Spliting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=24)

# One-hot encoding the target variable
y_train_encoded = to_categorical(y_train, num_classes=15)
y_test_encoded = to_categorical(y_test, num_classes=15)

# Defining Model
model_1 = keras.Sequential([
    layers.Dense(3, kernel_regularizer=regularizers.l2(0.001), activation='relu', input_shape=[x_train.shape[1]]),
    layers.Dense(24, kernel_regularizer=regularizers.l2(0.001), activation='relu'),
    layers.Dense(24, kernel_regularizer=regularizers.l2(0.001), activation='relu'),
    layers.Dense(16, kernel_regularizer=regularizers.l2(0.001), activation='relu'),
    layers.Dense(15)
])

# Compiling the model
optimizer = keras.optimizers.Adam(learning_rate=0.001)
loss_function = keras.losses.CategoricalCrossentropy(from_logits=True)

model_1.compile(loss=loss_function,
               optimizer=optimizer,
               metrics=['accuracy'])

# Trained the model for 2000 epochs
history = model_1.fit(x_train, y_train_encoded, epochs=2000, validation_data=(x_test, y_test_encoded), verbose=2)


# Plotted accuracy vs. epochs
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy vs. Epochs of Model-01')
plt.show()
```

```python
# Defined model:
model_2 = keras.Sequential([
    layers.Dense(3, kernel_regularizer=regularizers.l2(0.001), activation='relu', input_shape=[len(x_train.keys())]),
    layers.Dense(24, kernel_regularizer=regularizers.l2(0.001), activation='relu'),
    layers.Dense(24, kernel_regularizer=regularizers.l2(0.001), activation='relu'),
    layers.Dense(15)
])


# Compiled the model
optimizer = keras.optimizers.Adam(learning_rate=0.001)
loss_function = keras.losses.CategoricalCrossentropy(from_logits=True)

model_2.compile(loss=loss_function,
               optimizer=optimizer,
               metrics=['accuracy'])

# Trained the model for 2000 epochs
history = model_2.fit(x_train, y_train_encoded, epochs=2000, validation_data=(x_test, y_test_encoded), verbose=2)


# Plotted accuracy vs. epochs
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy vs. Epochs of Model-02')
plt.show()
```

```python
#Defined Model:
model_3 = keras.Sequential([
    layers.Dense(3, kernel_regularizer=regularizers.l2(0.001), activation='sigmoid', input_shape=[len(x_train.keys())]),
    layers.Dense(24, kernel_regularizer=regularizers.l2(0.001), activation='sigmoid'),
    layers.Dense(24, kernel_regularizer=regularizers.l2(0.001), activation='sigmoid'),
    layers.Dense(16, kernel_regularizer=regularizers.l2(0.001), activation='sigmoid'),
    layers.Dense(15)
])

# Compiled the model
optimizer = keras.optimizers.Adam(learning_rate=0.001)
loss_function = keras.losses.CategoricalCrossentropy(from_logits=True)

model_3.compile(loss=loss_function,
               optimizer=optimizer,
               metrics=['accuracy'])

# Trained the model for 2000 epochs
history = model_3.fit(x_train, y_train_encoded, epochs=2000, validation_data=(x_test, y_test_encoded), verbose=2)


# Plotted accuracy vs. epochs
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy vs. Epochs of Model-03')
plt.show()
```
